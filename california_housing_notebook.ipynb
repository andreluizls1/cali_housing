# California Housing Prices

Neste projeto serão realizados processos de Regressão Linear, a partir do dataset California Housing Prices que pode ser encontrado no [Kaggle](https://www.kaggle.com/datasets/camnugent/california-housing-prices) e disponibilizado por [Cam Nugent ](https://www.kaggle.com/camnugent) em 2018.

# 1. Contexto

Os dados referem-se às casas encontradas em um determinado distrito da Califórnia e a algumas estatísticas resumidas sobre elas com base nos dados do censo de 1990.

## 1.1. Objetivo

A partir dos dados meu objetivo é entender a relação entre as outras variáveis e o preço mediano dos imóveis, e fazer previsões de preço a partir desses parâmetros.

## 1.2. Variáveis

`longitude`: Uma medida do oeste de uma casa. Um valor mais negativo está mais a oeste.

`latitude`: Uma medida do norte mais alto de uma casa; um valor maior é mais ao norte.

`housing_median_age`: Idade mediana de uma casa em um bloco. Um número menor é um edifício mais recente.

`total_rooms`: Número total de salas em um bloco.

`total_bedrooms`: Número total de quartos em um bloco.

`population`: Número total de pessoas que residem em um bloco.

`households`: Número total de residências, um grupo de pessoas que moram em uma unidade residencial, para um bloco.

`median_income`: Renda mediana para famílias em um bloco de casas (medida em dezenas de milhares de dólares americanos).

`median_house_value`: Valor mediano da casa para casas em um bloco (medido em dólares americanos)

`ocean_proximity`: Próximidade do oceano.

# 5. Modelos de Regressão Linear

## 5.1. Pré processamento

### Definindo datasets de treino e teste

```python
# Importação do train_test_split

from sklearn.model_selection import train_test_split
```

```python
# Variável dependente

y = df['median_house_value']
```

```python
# Variáveis independentes

X = df.drop(['median_house_value'], axis = 1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 21)
```

## 5.2. Statsmodels

```python
# Importação de statsmodels

import statsmodels.api as sm
```

```python
# Adicionando a constante à X

X_train_constante = sm.add_constant(X_train)
```

```python
# fit do modelo

modelo_statsmodels = sm.OLS(y_train, X_train_constante, hasconst = True).fit()
```

```python
# Resumo do modelo

print(modelo_statsmodels.summary())
```

### Algumas conclusões a partir do summary()

* O R² de 64,5% aponta que o modelo tem um bom ajuste aos dados, após o R²-ajustado ele mantém a variabilidade de 64,5% apontando ainda um bom ajuste.
* O Prob (F-statistic) inferior a 0.05 mostra que o modelo tem resultados confiaveis.
* Todas as variáveis independentes tem um efeito significativo em median_house_value (Todas possuem P>|t| menor que 0.05).
* A idade mediana dos imóveis e a renda mediana são os fatores de maior peso na preço dos imóveis.
* Áreas mais populadas, com maior densidade populacional (número de quartos e total de residências) tem imóveis mais baratos.
* Não será necessário remover variáveis para prosseguir para o próximo passo.

## 5.3. Regressão Linear

```python
# Importação de LinearRegression e metrics

from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
# Criação da variável com o modelo

modelo = LinearRegression()
```

```python
# fit do modelo

modelo.fit(X_train, y_train)
```

### Coeficiente de determinação R²

```python
print('R² = {}'.format(modelo.score(X_train, y_train).round(4)))
```

### Gerando previsões para os dados de teste e seu R²

```python
y_previsto = modelo.predict(X_test)
```

```python
print('R² = {}'.format(metrics.r2_score(y_test, y_previsto).round(4)))
```

### Previsão pontual

```python
entrada = X_test[0:1]
entrada
```

```python
print('{0:.2f} USD'.format(modelo.predict(entrada)[0]))
```

### Gerando previsões

```python
# A partir da previsão pontual

longitude = -122.25
latitude = 37.77
housing_median_age = 52.0
total_rooms = 2650.0
total_bedrooms = 566.0
population = 1468.0
households = 567.0
median_income = 3.0161
ocean = 0
inland = 0
island = 0
nearbay = 1
nearocean = 0

entry = [[longitude, latitude, housing_median_age, total_rooms, total_bedrooms,
         population, households, median_income, ocean, inland, island, nearbay,
         nearocean]]

print('{0: .2f} USD'.format(modelo.predict(entry)[0]))
```

```python
# Alterando os parâmetros

longitude = -125
latitude = 38.5
housing_median_age = 15.5
total_rooms = 1930
total_bedrooms = 750
population = 1500
households = 350
median_income = 3.0161
ocean = 0
inland = 0
island = 0
nearbay = 0
nearocean = 1

entry_2 = [[longitude, latitude, housing_median_age, total_rooms, total_bedrooms,
         population, households, median_income, ocean, inland, island, nearbay,
         nearocean]]

print('{0: .2f} USD'.format(modelo.predict(entry_2)[0]))
```

### Métricas do modelo

```python
EQM = metrics.mean_squared_error(y_test, y_previsto).round(4)
REQM = np.sqrt(metrics.mean_squared_error(y_test, y_previsto)).round(4)
R2 = metrics.r2_score(y_test, y_previsto).round(4)

print('O modelo possui erro quadrático médio de {0:.2f}'.format(EQM))
print('O modelo possui raíz do erro quadrático médio de {0:.2f}'.format(REQM))
print('O modelo possui coeficiente de determinação de {0:.2f}%'.format(R2 * 100))
```
